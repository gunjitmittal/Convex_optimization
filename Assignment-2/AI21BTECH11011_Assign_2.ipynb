## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Function Defintion

```python
def f(x): # f(x)
    return 1-x**2*np.exp(-x)
```

```python
def f_(x): # f'(x)
    return x**2*np.exp(-x) - 2*x*np.exp(-x)
```

```python
def f__(x): # f''(x)
    return -x**2*np.exp(-x) + 4*x*np.exp(-x) - 2*np.exp(-x)
```

## Golden Section Search

```python
# Initialising the values
rho = float((3-5**0.5)/2)
l,u,eps = -0.5,2.5,0.05
w = float(u-l)
l_,u_ = l+w*rho,u-w*rho
fl_,fu_ = f(l_),f(u_)
width,minim = [],[]
```

```python
# Performing the search
while(w>eps):
    if fu_ > fl_:
        u = u_
        w = u - l
        u_,fu_ = l_,fl_
        l_ = l+w*rho
        fl_ = f(l_)
    else:
        l = l_
        w = u-l
        l_,fl_ = u_,fu_
        u_ = u-w*rho
        fu_ = f(u_)
    width.append(w)
    minim.append((l+u)/2)
```

```python
# Plotting the result
fig,axs = plt.subplots(1,2, constrained_layout=True)
axs[0].plot(width)
axs[0].set_ylabel("width")
axs[0].set_xlabel("Iterations")
axs[1].plot(minim)
axs[1].set_ylabel("minima")
axs[1].set_xlabel("Iterations")
fig.suptitle("Golden Section Search")
plt.show()
```

## Bisection Search

```python
# Initialising the values
l,u,eps = -0.5,2.5,0.05
w = u-l
x = 0 
width,minim = [],[]
```

```python
# Performing the search
while w>eps:
    c = (l+u)/2
    fc = f_(c)
    if fc>0:
        u = c
    elif fc<0:
        l = c
    else:
        x = c
        break
    w = u-l
    width.append(w)
    minim.append(c)
```

```python
# Plotting the result
fig,axs = plt.subplots(1,2, constrained_layout=True)
axs[0].plot(width)
axs[0].set_ylabel("width")
axs[0].set_xlabel("Iterations")
axs[1].plot(minim)
axs[1].set_ylabel("minima")
axs[1].set_xlabel("Iterations")
fig.suptitle("Bisection Search")
plt.show()
```

## Newton Search

```python
# Initialising the values
tail = -0.5
eps = 0.05
deri = []
```

```python
# Performing the search
while(np.abs(f_(tail))>eps):
    tail = tail - (f_(tail))/(f__(tail))
    deri.append(f_(tail))
```

```python
# Plotting the result
plt.plot(deri)
plt.title("Newton method")
plt.xlabel("Iterations")
plt.ylabel("Value of Derivative at tail")
plt.show()
```
