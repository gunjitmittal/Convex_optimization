# Q2
## Function Definitions

```python
def g(x):
    return (np.exp(-x)-np.cos(x))/np.exp(x)
```

```python
def g_(x):
    return (np.cos(x)+np.sin(x)-2*np.exp(-x))/np.exp(x)
```

```python
def g1(x):
    return np.cbrt(x)**5+30*x+20
```

```python
def g1_(x):
    return (5/3)*(np.cbrt(x)**2)+30
```

### g1(x)

```python
# Initialisations
head = -1.0
eps = 0.01
deri,x = [],[]
```

```python
# Newton raphson method
while(np.abs(g(head))>eps):
    x.append(head)
    deri.append(np.abs(g_(head)))
    head = head - g(head)/g_(head)
```

```python
# Plotting derivative vs iterations
fig,axs = plt.subplots(1,2, constrained_layout=True)
axs[0].plot(deri)
axs[0].set_ylabel("Derivative")
axs[0].set_xlabel("Iterations")
# Plotting x vs iterations
axs[1].plot(x)
axs[1].set_ylabel("x_k")
axs[1].set_xlabel("Iterations")
fig.suptitle("Newton Raphson g1")
plt.show()
```

## Function Definitions

```python
def g(x,n,a):
    return x**n-a
```

```python
def g_(x,n):
    return n*x**(n-1)
```

## Initialising variable

```python
number = int(input("Enter number"))
degree = int(input("Enter degree"))
head = 1
eps = 0.001
```

## Newton Raphson algo

```python
while(np.abs(g(head,degree,number))>eps):
    head = head - g(head,degree,number)/g_(head,degree)
print(head)
```
