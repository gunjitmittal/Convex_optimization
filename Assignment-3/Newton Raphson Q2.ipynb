```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
    return (np.exp(-x)-np.cos(x))/np.exp(x)
```

```python
def g_(x):
    return (np.cos(x)+np.sin(x)-2*np.exp(-x))/np.exp(x)
```

```python
def g1(x):
    return np.cbrt(x**5)+30*x+20;
```

```python
def g1_(x):
    return (5/3)*(np.cbrt(x**2))+30
```

```python
head = -1.0
eps = 0.05
deri,x = [],[]
i = 0 
```

```python
while(np.abs(g1(head))>eps and i<10):
    print(g1(head),g1_(head), head)
    x.append(head)
    deri.append(np.abs(g1_(head)))
    head = head - g1(head)/g1_(head)
    i+=1
```

```python
fig,axs = plt.subplots(1,2, constrained_layout=True)
axs[0].plot(deri)
axs[0].set_ylabel("Derivative")
axs[0].set_xlabel("Iterations")
axs[1].plot(x)
axs[1].set_ylabel("x_k")
axs[1].set_xlabel("Iterations")
fig.suptitle("Newton Raphson")
plt.show()
```
