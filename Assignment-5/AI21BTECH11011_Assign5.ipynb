## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Initializations

```python
def f(x,y):                     # The function
    return (2*x**2+x*y+y**2)
c = 50                          # The label value
drange = 50                     # The range on X and Y axis 
```

## Functions

```python
def plot_level(f,c,drange):
    """
    Takes as input function, level value and the range on x and y axis 
    and plots the level set coressponding to the above values. This 
    function makes a meshgrid of the x and y arrays so consumes some space
    but can do matrix operations on it so is fast
    """
    density = drange*50
    a=np.linspace(-drange,drange,density)
    b=np.linspace(-drange,drange,density)
    X,Y = np.meshgrid(a,b)
    eps=0.5
    truth = np.abs(f(X,Y)-c)<=eps
    points=np.vstack((X[truth],Y[truth]))
    plt.scatter(points[0,:],points[1,:])
```

```python
plot_level(f,c,drange)
```

```python
def plot_level2(f,c,drange):
    """
    Takes as input function, level value and the range on x and y axis 
    and plots the level set coressponding to the above values. This 
    function does not use meshgrid so takes less space but takes more time
    cause it uses a for loop.
    """
    density = drange*50
    a=np.linspace(-drange,drange,density)
    b=np.linspace(-drange,drange,density)
    eps=0.5
    for x in a:
        truth = np.abs(f(x,b)-c)<=eps
        plt.scatter(x*np.ones(b[truth].shape),b[truth],color='b')
    plt.show()
```

```python
plot_level2(f,c,drange)
```
